```python
import numpy as np
import matplotlib.pyplot as plt

def f(y):
    return 1/y

def sol(t):
    return (2*t)**0.5
```

```python
h = 0.01
t0 = 0.01
T = 2

t = np.linspace(t0, T, num=int(np.floor((T - t0)/h)))

true = sol(t)

tk = t0
y0 = sol(t0)
est_vals = np.zeros(np.shape(t))
est_vals[0] = y0
for i in range(len(est_vals)-1):
    est_vals[i+1] = est_vals[i] + h*f(est_vals[i])
    
```

```python
plt.scatter(t, true)
plt.scatter(t, est_vals, c='r')
plt.show()
```
